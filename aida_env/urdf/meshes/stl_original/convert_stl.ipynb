```python
from stl import mesh
from mpl_toolkits import mplot3d
from matplotlib import pyplot
import math
%matplotlib inline
```

```python

def plot(mesh):
    # Create a new plot
    figure = pyplot.figure()
    axes = mplot3d.Axes3D(figure)

    axes.set_xlim3d(-0.200, 0.200)
    axes.set_ylim3d(-0.200,0.200)
    axes.set_zlim3d(-0.200,0.200)

    axes.add_collection3d(mplot3d.art3d.Poly3DCollection(mesh.vectors))

    # Show the plot to the screen
    pyplot.show()
```

```python


lower_leg = mesh.Mesh.from_file('lower_leg.stl')
lower_leg.translate([-420, 0, 50])
lower_leg.x *= 0.001
lower_leg.y *= 0.001
lower_leg.z *= 0.001
lower_leg.rotate([0.,1.0,0.], math.radians(90))
lower_leg.z *=0.4/0.5
```

```python
plot(lower_leg)
lower_leg.save("../walking_base_v0/lower_leg.stl")
```

```python
upper_leg = mesh.Mesh.from_file('upper_leg.stl')
upper_leg.x *= 0.001
upper_leg.y *= 0.001
upper_leg.z *= 0.001
upper_leg.rotate([0.,0.,1.0], math.radians(-90))
upper_leg.translate([0, -0.2, 0.])

plot(upper_leg)
upper_leg.save("../walking_base_v0/upper_leg.stl")
```

```python
chassis = mesh.Mesh.from_file('chassis.stl')
chassis.x *= 0.001
chassis.y *= 0.001
chassis.z *= 0.001
#chassis.rotate([0.,0.,1.0], math.radians(90))
chassis.translate([0, 0, -0.1])

plot(chassis)
chassis.save("../walking_base_v0/chassis.stl")
```

```python
shoulder = mesh.Mesh.from_file('shoulder.stl')
shoulder.x *= 0.001
shoulder.y *= 0.001
shoulder.z *= 0.001
shoulder.rotate([0.,0.,1.0], math.radians(90))
shoulder.rotate([0.,1.0,0.], math.radians(90))
shoulder.translate([0, 0.02, 0.])

plot(shoulder)
shoulder.save("../walking_base_v0/shoulder.stl")
```
